```python
#Merge,join,and concatenate
```

```python
data1 = {"Emp id": ['E01','E02','E09','E08','E05','E06'],
        "Name":["Ram","shyam","Rahul","vishal","Ravi",'john'],
        "Age":[45,34,26,27,20,39]}

data2 = {"Emp id": ['E01','E02','E03','E04','E05','E06'],
        "Salary":[30000,34000,45000,35000,38000,58000]}
df6 = pd.DataFrame(data1)
df7 = pd.DataFrame(data2)
```

```python
pd.merge(df6,df7,on='Emp id',how ='inner')
```

```python
pd.merge(left=df6,right=df7,on='Emp id',how='left')
```

```python
pd.merge(left=df6,right=df7,on='Emp id',how='right')
```

```python
Dict = {"Fruits":["apple","banana","Grapes","mango"],
       "price":[100,200,150,200],
       "Quantity":[15,20,19,10]}

df1 = pd.DataFrame(Dict)
print(df1)
df2 = df1.copy()
df2
df2.loc[0,"price"] = 120
df2.loc[1,"price"] = 175
df2.loc[3,"price"] = 30
df2.loc[0,"Quantity"] = 120
df2.loc[1,"Quantity"] = 15
df2.loc[3,"Quantity"] = 5
```

```python
print(df1.compare(df2))
```

```python
print(df1.compare(df2,keep_shape=True))
```

```python
# pivoting and melting 
```

```python
Dict = {'keys':['K1','K2','K1','K2'],
       "Name":['john','ben','David','Peter'],
       "House":['red',"Blue",'green','red'],
       "Grade":['3rd','8th','9th','8th']}
df = pd.DataFrame(Dict)
print(df)
```

```python
print(df.pivot(index='keys',columns='Name',values=['House','Grade']))
```

```python
print(pd.melt(df,id_vars=['Name'],value_vars=['House']))
```
